```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import *
x = np.arange(-20, 20, 0.2)
y = (3*np.pi/2)*np.sin(x)/x+np.cos(x)
plt.plot(x,y)
plt.axhline(y=0)#x軸
plt.axhline(y=1)#x軸
plt.axhline(y=-1)#x軸
plt.axvline(x=0)#y軸
plt.show()
```

```python
import numpy as np

A = np.array([
    [0, 1, 0, 0],
    [1, 0, np.sqrt(2), 0],
    [0, np.sqrt(2), 0, np.sqrt(3)],
    [0, 0, np.sqrt(3), 0]
])
B=np.array([
    [0, -1, 0, 0],
    [1, 0, -np.sqrt(2), 0],
    [0, np.sqrt(2), 0, -np.sqrt(3)],
    [0, 0, np.sqrt(3), 0]
])
C=B.dot(A)
D=A.dot(B)
print(D-C)
print()
```
