# Fabey Perot in meep(AB)

## first of all load the library

```python
import meep as mp
import numpy as np
import matplotlib.pyplot as plt
```

## 定義參數

```python
fmin = 1.5 #300THz
fmax = 3.5 #600THz
nf = 1000 
fcen = (fmin + fmax)/2
df = fmax - fmin
#dfp = 1/2
fre = np.linspace(fmin,fmax,nf)
omega = 2 * np.pi * fre
```

```python
dpml = 1

dA = 0.2
dB = 0.3
epsilonA = 12
epsilonB = 3

NAB = 15

sz = 2*dpml+16

resolution = 100 # pixels/μm

cell_size = mp.Vector3(z=sz)

pml_layers = [mp.PML(dpml)]

source_z=-sz/2+dpml
flux_reg_z=sz/2-dpml
```

## Sources set up

```python
sources = [mp.Source(mp.GaussianSource(fcen,fwidth=df),
                     component=mp.Ex,
                     center=mp.Vector3(z=source_z),
                     size=mp.Vector3())]
```

## Simulation 真空

```python
sim_vac = mp.Simulation(cell_size=cell_size,
                    resolution=resolution,
                    boundary_layers=pml_layers,
                    dimensions=1,
                    sources=sources)
refl_fr = mp.FluxRegion(center=mp.Vector3(z=-8))
refl = sim_vac.add_flux(fcen, df,nf, refl_fr)
tran_fr = mp.FluxRegion(center=mp.Vector3(z=8))
tran = sim_vac.add_flux(fcen, df, nf, tran_fr)
sim_vac.run(until_after_sources=mp.stop_when_fields_decayed(50,mp.Ex,
                                                            mp.Vector3(z=0),1e-3))
vac_flux = mp.get_fluxes(tran)
flux_freqs = mp.get_flux_freqs(tran)
```

```python
plt.figure(dpi=150)#解析度
plt.plot(omega/2/np.pi,T,'-',label='$T_{nur}$')#x軸omega y軸T 
plt.xlabel("omega ($2\pi$/μm)")
plt.ylabel("$T$")
plt.xlim([fmin,fmax])
plt.ylim([0,1])
plt.savefig('AB.png')
plt.show()
```
