```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ml = np.random.normal(0,1,Nt)
nl = np.random.normal(0,1,Nt)
```

```python
def Dnsqt(wla,T):
    hdk=7.63823258e-12
    return np.sqrt(6*hdk**2*abs(wla)/(np.exp(hdk*abs(wla)/T)-1)/T**2/np.pi)
```

```python
M=2**21
tsim=7e-12
eps0 = 8.854187e-12
hbar = 1.05457182e-34
c = 2.99792458e8
kb = 1.380649e-23

deltap = np.sqrt(1/3/eps0/hbar/c)*kb
resolution = 50 
tsim = 1000
courant = 0.5 
dt = courant  / resolution
Nt = int(tsim/dt + 1)
tslice = np.linspace(0, 2e-15, Nt)
omegal = 2 * np.pi / Nt * np.linspace(0,Nt-1,Nt) 

dlsqr = np.empty(Nt)
for i in range(Nt):
     dlsqr[i] = Dnsqt(omegal[i],30000)

fc = (ml+1j*nl)*dlsqr

def Es(tj,T):
    hdk=7.63823258e-12
    #Es0 = rand()*np.sqrt(6*hdk/np.pi/T)
    Esm = 0
    for j in range(-M,M-1,1):
        wl = 2*np.pi*j/tsim
        if j == 0:
            Es0 = ml()*np.sqrt(6*hdk**2/np.pi/T)
        else:
            Esm = Esm + (ml()*np.cos(wl*tj)-nl()*np.sin(wl*tj))*Dnsqt(wl,T)
    return deltap*T*(Es0 + Esm)/np.sqrt(tsim)
```

```python

tl = np.linspace(0,2e-15,200)
T = 50000
plt.plot(tl,Es(tl,T),'.',label=50000)
tl = np.linspace(0,2e-15,200)
T = 30000
Es = Es(tl,T)
plt.plot(tl,Es,'.',label=30000)
plt.xlabel('Time')
plt.ylabel('Es(V/m)')
plt.grid()
plt.legend()
```

```python
dft = np.fft.fft(Es)

# Compute the frequencies associated with each DFT coefficient
freqs = np.linspace(0,25*1e15,200)

# Plot the results
import matplotlib.pyplot as plt
plt.plot(freqs,abs(dft),'.')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.show()
```
